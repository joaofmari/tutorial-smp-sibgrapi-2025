```python
import os
import shutil
import numpy as np
from PIL import Image
```

```python
train_path = os.path.expanduser('~/Datasets/38-Cloud/38-Cloud_training/')
```

```python
images_red_list = sorted([f for f in os.listdir(os.path.join(train_path, "train_red")) if f.endswith('.TIF')])
images_green_list = sorted([f for f in os.listdir(os.path.join(train_path, "train_green")) if f.endswith('.TIF')])
images_blue_list = sorted([f for f in os.listdir(os.path.join(train_path, "train_blue")) if f.endswith('.TIF')])
images_nir_list = sorted([f for f in os.listdir(os.path.join(train_path, "train_nir")) if f.endswith('.TIF')])
images_gt_list = sorted([f for f in os.listdir(os.path.join(train_path, "train_gt")) if f.endswith('.TIF')])

print(len(images_red_list))
```

```python
kept = 0
removed = 0

filename = os.path.join(train_path, 'training_patches_38-Cloud.csv')
backup = os.path.join(train_path, 'training_patches_38-Cloud_ORIGINAL.csv')

# Copy only if the backup file does not exist
if not os.path.exists(backup):
    shutil.copy(filename, backup)

file = open(filename, 'w')
file.write('name\n')

for idx in range(len(images_red_list)):
    print(f'\nProcessing image {idx}: {os.path.join(train_path, 'train_red', images_red_list[idx])}...')

    path_red = os.path.join(train_path, 'train_red', images_red_list[idx])
    path_green = os.path.join(train_path, 'train_green', images_green_list[idx])
    path_blue = os.path.join(train_path, 'train_blue', images_blue_list[idx])
    path_nir = os.path.join(train_path, 'train_nir', images_nir_list[idx])
    
    img_rgb = np.stack([np.array(Image.open(path_red)),
                        np.array(Image.open(path_green)),
                        np.array(Image.open(path_blue)),
                        np.array(Image.open(path_nir)),], axis=2)
    
    print(img_rgb.ndim, img_rgb.shape)

    H, W = img_rgb.shape[:2]
    total_pixels = H * W
    print(H, W)

    black_pixels = np.all(img_rgb == 0, axis=-1)

    num_black = np.count_nonzero(black_pixels)
    frac_black = num_black / total_pixels

    print(num_black, total_pixels, frac_black)
    if frac_black == 1.0:
        print(f'---> Image will be REMOVED!')
        removed += 1
        # To actually remove the images, uncomment the code below.
        # ---------------------------------
        # shutil.move(path_red, os.path.join(train_path, 'train_red_black'))
        # shutil.move(path_green, os.path.join(train_path, 'train_green_black'))
        # shutil.move(path_blue, os.path.join(train_path, 'train_blue_black'))
        # shutil.move(path_nir, os.path.join(train_path, 'train_nir_black'))
        # ---------------------------------

    else:
        print(f'---> Image will be KEPT!')
        kept +=1
        file.write(os.path.splitext(images_red_list[idx])[0].split("_", 1)[-1] + '\n')

file.close
```

```python
print('Kept: ', kept)
print('Removed: ', removed)
```
